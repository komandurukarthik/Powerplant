##### Class imbalance issue

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

Model building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Finding best random state

```python
from sklearn.tree import DecisionTreeClassifier
maxAccu=0
maxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print("Best Accuracy is:",maxAccu,"on Random state:",maxRs)
```

Best accuracy is 90% with a random state of 174

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRs)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
pred=DTC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

DTC Classifier gives 89% accuracy

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
pred=RFC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

Random forest classifier gives 93% Accuracy

```python
sv=SVC()
sv.fit(x_train,y_train)
pred=sv.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

svc gives 80% accuracy

Upon reviewing the above outputs the Random forest classifier gives 93% accuracy.So to choose the best model we will use cross validation Score.

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(DTC,x,y,cv=5).mean())
```

Decision tree gives 86% score

```python
print(cross_val_score(RFC,x,y,cv=5).mean())
```

Cross validation gives 90% accuracy

```python
print(cross_val_score(sv,x,y,cv=5).mean())
```

support vector gives 74% score

So our random forest classifier gives the less difference so we will consider this a best model
